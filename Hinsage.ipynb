#### ***GraphSAGE là một khung công cụ cho việc học biểu diễn dựa vào tiên đề trên đồ thị lớn. GraphSAGE được sử dụng để tạo ra các biểu diễn vector chiều thấp cho các nút, và đặc biệt hữu ích cho các đồ thị có thông tin thuộc tính nút phong phú***

## Hinsage:

*   Mở rộng thuật toán GraphSAGE cho các mạng không đồng nhất:các đồ thị chứa nhiều loại nút và cạnh khác nhau
* Hỗ trợ học biểu diễn, phân loại/hồi quy nút và dự đoán/hồi quy liên kết cho các biểu đồ không đồng nhất.
* Việc triển khai hiện tại hỗ trợ tổng hợp trung bình các nút lân cận, có tính đến loại của chúng và loại liên kết giữa chúng.

* Các loại quan hệ: (Nút bắt đầu- cạnh- nút kết thúc)
* Sử dụng các loại quan hệ để mô tả các cạnh
* Loại quan hệ tốt hơn để mô tả sự tương tác giữa các nút và các cạnh

image.png

* Các đồ thị không đồng nhất như: Đồ thị Kiến thức Y sinh, Đồ thị Sự kiện,Đồ thị Thương mại điện tử, Đồ thị học thuật

* Có thể coi các loại nút và cạnh là các đặc điểm.
* heterogeneous graph có thể coi như là:standard graph

#### Cần sử dụng Hinsage khi nào:

* Trường hợp 1: Các loại nút/cạnh khác nhau có các hình dạng đặc điểm khác nhau
* Trường hợp 2: Chúng ta biết rằng các loại quan hệ khác nhau đại diện cho các loại tương tác khác nhau

* Cuối cùng, đồ thị không đồng nhất là một biểu diễn đồ thị phong phú hơn
** Nắm bắt các loại tương tác khác nhau giữa các thực thể
* Tuy nhiên, nó cũng đi kèm với những chi phí
** Đắt đỏ hơn (tính toán, lưu trữ)
** Triển khai phức tạp hơn
* Có nhiều cách để chuyển đổi một đồ thị không đồng nhất thành một đồ thị tiêu chuẩn(nghĩa là một đồ thị đồng nhất)
